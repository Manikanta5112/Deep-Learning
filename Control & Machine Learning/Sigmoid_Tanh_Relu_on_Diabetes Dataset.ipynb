```python
def draw_loss_vs_acc(history):
    # Get training and test loss histories
    training_loss = history.history['accuracy']
    test_loss = history.history['val_accuracy']
    
    # Create count of the number of epochs
    epoch_count = range(1, len(training_loss) + 1)
    
    # Visualize loss history
    plt.plot(epoch_count, training_loss, 'r--')
    plt.plot(epoch_count, test_loss, 'b-')
    plt.legend(['Training accuracy', 'Test accuracy'])
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.show();
    
```

## Sigmoid

```python
# Load dependencies
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Load data
dataset = np.loadtxt('./diabetes.csv', delimiter=',')

# Separate input data and target data
X = dataset[:, 0:8] # taking all rows and columns from 0 to 7
Y = dataset[:, 8] # taking all rows and only 8th column

# Create the Model
model = Sequential()

# Model layers
model.add(Dense(12, input_shape=(8,), activation='sigmoid'))
model.add(Dense(8, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

# Model compilation
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Model training
sigmoid_history = model.fit(X, Y, epochs=225, batch_size=25, verbose=1, validation_split=0.2)
```

```python
draw_loss_vs_acc(sigmoid_history)
```

## tanh

```python
# Create the Model
model = Sequential()

# Model layers
model.add(Dense(12, input_shape=(8,), activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))

# Model compilation
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Model training
tanh_history = model.fit(X, Y, epochs=225, batch_size=25, verbose=1, validation_split=0.2)
```

```python
draw_loss_vs_acc(tanh_history)
```
